## 3.Graph Replication and Enhancement
---

<span style="font-size: larger;">**3.1 Reproduction of the Graph**</span></br>

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['18-34', '35-54', '55+', 'Total', '18-34', '35-54', '55+', 'Total']
impact = [62, 51, 38, 51, 38, 49, 62, 50]
colors_darker_gray = ['#404040', 'blue', 'lightblue', 'orange', '#404040', 'blue', 'lightblue', 'orange']
max_height = max(impact)

# Plot the bars
fig, ax = plt.subplots(figsize=(10, 6))
for idx, cat in enumerate(categories):
    bar_color = colors_darker_gray[idx]
    bar = ax.bar(idx, impact[idx], color=bar_color)
    height = bar[0].get_height()
    ax.text(bar[0].get_x() + bar[0].get_width() / 2, height + 1, f'{impact[idx]}%', ha='center', va='bottom', color=bar_color)

# Add labels over the bars
ax.text(1.5, max_height + 3, 'A great deal or Somewhat', ha='center', va='bottom', fontsize=14, color='gray')
ax.text(5.5, max_height + 3, 'A little or Not at all', ha='center', va='bottom', fontsize=14, color='gray')

# Set the x-axis
ax.set_xticks(range(len(categories)))
ax.set_xticklabels(categories)

# Set the y-axis
ax.set_yticks([0, 10, 20, 30, 40, 50, 60])
ax.set_yticklabels(['0%', '10%', '20%', '30%', '40%', '50%', '60%'])

# Draw a vertical line to separate
plt.axvline(x=3.5, color='gray', linestyle='--')

# Create the horizontally arranged legend
legend_labels = ['18-34', '35-54', '55+', 'Total']
legend_colors = colors_darker_gray[:4]
circles = [plt.Line2D([0], [0], marker='o', color='w', label=lbl, markersize=20, markerfacecolor=clr) for lbl, clr in zip(legend_labels, legend_colors)]
legend = ax.legend(handles=circles, loc='upper left', bbox_to_anchor=(-0.047, 1.13), ncol=len(circles))

# Set the title
ax.set_title('Expected climate change impact by age', x=0.23 ,y=1.13, color='gray', fontsize=16, fontweight='bold')

# Adding the right frame line and removing the top frame line
ax.spines['right'].set_visible(True)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

<span style="font-size: larger;">**3.2 Shortcomings of the Graph**</span></br>
- The comparison is not intuitive. The artwork uses A vertical line to separate A greate deal or Somewhat from A little or Not at all. Readers need to cross several bars to compare the two types of content, which causes difficulties in reading. In addition, this kind of correspondence uses the same color. This does not show the difference between the two categories.
- Repeat. Both the legend and the content on the X-axis represent the age group, which has the same meaning and redundant representation, which is not conducive to readers' reading.
- The drawing does not conform to the specification. There are many places in the diagram that violate the specification of the drawing. For example, the X axis and Y axis of the figure do not have labels, so readers cannot directly understand the inner meaning through the figure, and need to understand through additional words.
- There is an error in the data in the graph. The value in the total column for 'A geate deal or somewhat' should be 50%.

<span style="font-size: larger;">**3.3 Improvement Suggestions with Beneficial Changes for the Graph**</span></br>
- Convert the chart to a grouped bar chart. Two bars of complementary colors are used for each age group to represent A great deal or Somewhat and A little or Not at all. In this way, readers can intuitively identify and compare the differences between different age groups and reduce visual confusion.
- Use axes to represent age groups and legends to represent categories. Such adjustments can clearly distinguish data categories and age groups, avoid visual duplication, and simplify the interpretation of charts.
- Perfect drawing details. Add explicit labels on the X and Y axes, and add a descriptive title to the chart. In addition, it is also necessary to adjust the contents of the chart frame and line to present a good reading effect to the reader.

<span style="font-size: larger;">**3.4 Display the Modified Graph**</span>

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['18-34', '35-54', '55+', 'Total']
impact_great_deal = [62, 51, 38, 50]
impact_little_none = [38, 49, 62, 50]

# Set the width of a bar
bar_width = 0.2
# Positions of the left bar-boundaries
bar_l = [i for i in range(len(categories))]
# Positions of the right bar-boundaries
bar_r = [i + bar_width for i in bar_l]
fig, ax = plt.subplots(figsize=(8, 6))
# Plot the bars for "A great deal or Somewhat"  and "A little or Not at all" 
bars_left = ax.bar(bar_l, impact_great_deal, width=bar_width, color='#404040', label='A great deal or Somewhat')
for idx, bar in enumerate(bars_left):
    ax.text(bar.get_x() + bar.get_width()/2, impact_great_deal[idx]+0.2 , f'{impact_great_deal[idx]}%', horizontalalignment='center', verticalalignment='bottom')
bars_right = ax.bar(bar_r, impact_little_none, width=bar_width, color='#FF8C00', label='A little or Not at all')
for idx, bar in enumerate(bars_right):
    ax.text(bar.get_x() + bar.get_width()/2, impact_little_none[idx]+0.2, f'{impact_little_none[idx]}%', horizontalalignment='center', verticalalignment='bottom')

# Set the x-axis
ax.set_xlabel('Age', fontsize=12)
ax.set_xticks([r + bar_width/2 for r in bar_l])
ax.set_xticklabels(categories)

# Set y-axis
ax.set_ylabel('Population proportion', fontsize=12)
ax.set_yticks([0, 10, 20, 30, 40, 50, 60, 70])
ax.set_yticklabels(['0%', '10%', '20%', '30%', '40%', '50%', '60%', '70%'])

# Set the title
ax.set_title('Expected climate change impact by age', y=1.05, color='black', fontsize=16, fontweight='bold')
# Set the legend
ax.legend(bbox_to_anchor=(1.04, 1.03))

# Remove the background grid
ax.grid(False)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
